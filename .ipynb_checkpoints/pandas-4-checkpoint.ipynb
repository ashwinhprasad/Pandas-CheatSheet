## Aggregate

```python
df = pd.DataFrame([[1, 2, 3],
                   [4, 5, 6],
                   [7, 8, 9],
                   [np.nan, np.nan, np.nan]],
                  columns=['A', 'B', 'C'])
df.head()
```

```python
df.agg(['mean','sum'])
```

```python
df.agg({ 'A':'sum', 'B': 'min'})
```
